```python
# 2.1   理论计算题

import numpy as np

# 定义向量和矩阵
a = np.array([2, -1, 3])
b = np.array([1, 4, -2])
A = np.array([[1, 0, 2], [-1, 3, 1]])
B = np.array([[2, 1], [0, -1], [3, 2]])

# 1. 向量点积
dot_product = np.dot(a, b)

# 2. 矩阵乘法 A * B
matrix_multiplication = np.dot(A, B)

# 3. 向量 a 的 Frobenius 范数
frobenius_norm = np.linalg.norm(a)

# 输出结果
print("向量点积 a · b:", dot_product)
print("矩阵乘法 A × B:", matrix_multiplication)
print("向量 a 的 Frobenius 范数:", frobenius_norm)
```

```python
# 3.1   理论计算题

# 给定的条件
P_disease = 0.001  # 患病率
P_positive_given_disease = 0.99  # 灵敏度
P_positive_given_no_disease = 0.02  # 假阳性率
P_no_disease = 1 - P_disease  # 未患病率

# 计算 P(阳性)
P_positive = P_positive_given_disease * P_disease + P_positive_given_no_disease * P_no_disease

# 使用贝叶斯公式计算 P(病|阳性)
P_disease_given_positive = (P_positive_given_disease * P_disease) / P_positive

print("真正患病的概率 P(病|阳性):", P_disease_given_positive)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats

# 设置支持中文的字体
plt.rcParams['font.sans-serif'] = ['SimHei']  # 使用黑体字体
plt.rcParams['axes.unicode_minus'] = False  

# 设置参数
n = 10000  # 每次生成的随机数个数
m = 1000  # 重复次数

# 生成均匀分布的随机变量并计算均值
means = [np.mean(np.random.uniform(0, 1, n)) for _ in range(m)]

# 绘制直方图
plt.hist(means, bins=30, density=True, alpha=0.6, color='g', label="样本均值的分布")

# 绘制理论正态分布曲线
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, np.mean(means), np.std(means))
plt.plot(x, p, 'k', linewidth=2, label="理论正态分布")

# 计算均值的方差
variance = np.var(means)

# 输出结果
plt.legend()
plt.title("均值的直方图与理论正态分布")
plt.show()

print("均值的实际方差:", variance)
```

```python
# 5.2 编程题

import numpy as np
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
import matplotlib.pyplot as plt

# 1. 加载数据并进行预处理
digits = load_digits()
X = digits.data
y = digits.target

# 进行one-hot编码
encoder = OneHotEncoder(sparse_output=False)  # 使用 sparse_output 替代 sparse
y_one_hot = encoder.fit_transform(y.reshape(-1, 1))

# 数据集分割
X_train, X_test, y_train, y_test = train_test_split(X, y_one_hot, test_size=0.2, random_state=42)

# 2. 实现softmax函数
def softmax(z):
    return np.exp(z) / np.sum(np.exp(z), axis=1, keepdims=True)

# 3. 定义交叉熵损失函数
def cross_entropy_loss(y_true, y_pred):
    m = y_true.shape[0]
    return -np.sum(y_true * np.log(y_pred)) / m

# 4. 小批量随机梯度下降
def train_softmax(X, y, learning_rate=0.1, epochs=50, batch_size=32):
    m, n = X.shape
    k = y.shape[1]  # 输出类别数
    W = np.random.randn(n, k)
    b = np.zeros((1, k))

    loss_history = []  # 用于存储每个epoch的损失
    train_accuracy_history = []  # 用于存储每个epoch的训练准确率
    test_accuracy_history = []  # 用于存储每个epoch的测试准确率

    for epoch in range(epochs):
        # Mini-batch gradient descent
        for i in range(0, m, batch_size):
            X_batch = X[i:i + batch_size]
            y_batch = y[i:i + batch_size]

            # 前向传播
            z = np.dot(X_batch, W) + b
            y_pred = softmax(z)

            # 计算梯度
            dz = y_pred - y_batch
            dW = np.dot(X_batch.T, dz) / batch_size
            db = np.sum(dz, axis=0, keepdims=True) / batch_size

            # 更新参数
            W -= learning_rate * dW
            b -= learning_rate * db

        # 每50次迭代打印一次损失
        loss = cross_entropy_loss(y_batch, y_pred)
        loss_history.append(loss)

        # 计算训练集和测试集的准确率
        y_pred_train = predict(X_train, W, b)
        y_pred_test = predict(X_test, W, b)
        train_accuracy = np.mean(np.argmax(y_train, axis=1) == y_pred_train)
        test_accuracy = np.mean(np.argmax(y_test, axis=1) == y_pred_test)

        train_accuracy_history.append(train_accuracy)
        test_accuracy_history.append(test_accuracy)

        print(f'Epoch {epoch}, Loss: {loss:.4f}, Train Accuracy: {train_accuracy * 100:.2f}%, Test Accuracy: {test_accuracy * 100:.2f}%')

    return W, b, loss_history, train_accuracy_history, test_accuracy_history

# 训练模型
W, b, loss_history, train_accuracy_history, test_accuracy_history = train_softmax(X_train, y_train)

# 5. 测试模型
def predict(X, W, b):
    z = np.dot(X, W) + b
    y_pred = softmax(z)
    return np.argmax(y_pred, axis=1)

y_pred_train = predict(X_train, W, b)
y_pred_test = predict(X_test, W, b)

# 计算准确率
train_accuracy = np.mean(np.argmax(y_train, axis=1) == y_pred_train)
test_accuracy = np.mean(np.argmax(y_test, axis=1) == y_pred_test)

print(f'Training Accuracy: {train_accuracy * 100:.2f}%')
print(f'Test Accuracy: {test_accuracy * 100:.2f}%')

# 6. 可视化
# 绘制损失曲线
plt.figure(figsize=(10, 6))
plt.plot(loss_history, label="Loss", color="blue")
plt.title("Training Loss over Epochs")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.show()

# 绘制训练集和测试集的准确率曲线
plt.figure(figsize=(10, 6))
plt.plot(train_accuracy_history, label="Train Accuracy", color="green")
plt.plot(test_accuracy_history, label="Test Accuracy", color="red")
plt.title("Accuracy over Epochs")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 6.2   编程题

import numpy as np
import matplotlib.pyplot as plt

# 1. 生成数据
np.random.seed(42)
X_class_1 = np.random.randn(200, 2) + np.array([1, 1])
X_class_2 = np.random.randn(200, 2) + np.array([-1, -1])
X = np.vstack([X_class_1, X_class_2])
y = np.hstack([np.zeros(200), np.ones(200)])

# 2. 定义sigmoid函数和损失函数
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def binary_crossentropy_loss(y_true, y_pred):
    m = y_true.shape[0]
    return -np.mean(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))

# 3. 梯度下降法
def train_logistic_regression(X, y, learning_rate=0.1, epochs=1000):
    m, n = X.shape
    W = np.zeros(n)
    b = 0

    for epoch in range(epochs):
        # 计算预测值
        z = np.dot(X, W) + b
        y_pred = sigmoid(z)

        # 计算梯度
        dW = np.dot(X.T, (y_pred - y)) / m
        db = np.sum(y_pred - y) / m

        # 更新参数
        W -= learning_rate * dW
        b -= learning_rate * db

        if epoch % 100 == 0:
            loss = binary_crossentropy_loss(y, y_pred)
            print(f'Epoch {epoch}, Loss: {loss}')

    return W, b

# 训练模型
W, b = train_logistic_regression(X, y)

# 4. 绘制决策边界
def plot_decision_boundary(X, y, W, b):
    plt.figure(figsize=(10, 6))
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
    ax = plt.gca()
    x_min, x_max = ax.get_xlim()
    y_min, y_max = ax.get_ylim()

    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))
    Z = sigmoid(np.dot(np.c_[xx.ravel(), yy.ravel()], W) + b)
    Z = Z.reshape(xx.shape)
    ax.contour(xx, yy, Z, levels=[0.5], linewidths=2, colors='black')

    plt.show()

plot_decision_boundary(X, y, W, b)

# 5. 计算准确率
def calculate_accuracy(X, y, W, b):
    y_pred = sigmoid(np.dot(X, W) + b)  # 计算预测值
    y_pred_class = (y_pred >= 0.5).astype(int)  # 将概率转为类别
    accuracy = np.mean(y_pred_class == y)  # 计算准确率
    return accuracy

# 计算训练集和测试集的准确率
train_accuracy = calculate_accuracy(X, y, W, b)
print(f'Training Accuracy: {train_accuracy * 100:.2f}%')

# 使用新生成的100个样本作为测试集
X_test_class_1 = np.random.randn(100, 2) + np.array([1, 1])
X_test_class_2 = np.random.randn(100, 2) + np.array([-1, -1])
X_test = np.vstack([X_test_class_1, X_test_class_2])
y_test = np.hstack([np.zeros(100), np.ones(100)])

# 计算测试集准确率
test_accuracy = calculate_accuracy(X_test, y_test, W, b)
print(f'Test Accuracy: {test_accuracy * 100:.2f}%')
```
